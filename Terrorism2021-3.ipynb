```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
pd.set_option('display.max_columns', 200)
```

```python
df = pd.read_csv('globalterrorismdb_Jan2021-June2021.csv')
```

```python
df1 = pd.read_csv('globalterrorismdb_0522dist.csv')
```

```python
# Keep columns being used in analysis

columns_to_keep = ['iyear', 'imonth', 'iday',
        'country', 'country_txt', 'region', 'region_txt',
       'provstate', 'city', 'latitude', 'longitude',
        'multiple', 'success', 'attacktype1','attacktype1_txt', 'targtype1', 'targtype1_txt',
        'targsubtype1', 'targsubtype1_txt', 'target1', 'gname', 'weaptype1', 'weaptype1_txt',
        'weapsubtype1', 'weapsubtype1_txt','nkill', 'nwound', 'propextent', 'propextent_txt', 'propcomment']
```

```python
df = df[columns_to_keep].copy()

df1 = df1[columns_to_keep].copy()

df2 = df1[columns_to_keep].copy()
```

```python
df.describe()
```

```python
df.head(3)
```

```python
df.rename(columns={'iyear': 'year',
                   'imonth': 'month',
                   'iday':'day',
                   'gname': 'terrorist_group'
                   }, inplace=True)

df1.rename(columns={'iyear': 'year',
                   'imonth': 'month',
                   'iday':'day',
                   'gname': 'terrorist_group'
                   }, inplace=True)


df2.rename(columns={'iyear': 'year',
                   'imonth': 'month',
                   'iday':'day',
                   'gname': 'terrorist_group'
                   }, inplace=True)
# For Comparison: Filter 'df1' Dataframe to only include Afghanistan and Years[2019,2020]

df1 = df1[ df1['country_txt'] == 'Afghanistan']

df2 = df2[ df2['country_txt'] == 'Afghanistan']

df1 = df1[df1['year'].isin([2020])]

df2 = df2[df2['year'].isin([2019])]
```

```python



                        
    
                                # Exploratory Data Analysis



    
    
    
    

```

```python
#Dropping duplicates

df.drop_duplicates(inplace=True)
df1.drop_duplicates(inplace=True)
df2.drop_duplicates(inplace=True)
```

```python
# Combine [year, month, day] columns into single column called 'date'

df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
df1['date'] = pd.to_datetime(df1[['year', 'month', 'day']])
df2['date'] = pd.to_datetime(df2[['year', 'month', 'day']])
```

```python
# Drop columns

df.drop(columns=['year', 'month', 'day'], inplace=True)

df1.drop(columns=['year', 'month', 'day'], inplace=True)

df2.drop(columns=['year', 'month', 'day'], inplace=True)

# Move 'date' column to the front

column_to_move = 'date'
df.insert(0, column_to_move, df.pop(column_to_move))
df1.insert(0, column_to_move, df1.pop(column_to_move))
df2.insert(0, column_to_move, df2.pop(column_to_move))
```

```python
    # Bar plots showing 'Top 10 Dates With Most Terrorist Attacks' and 'Terrorist Attacks by Month'


# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))

# First, we'll extract the date part so [hour, minute, second] doesnt show
df['date_only'] = df['date'].dt.date

top_dates = df['date_only'].value_counts().head(10)
top_dates \
    .plot(kind='bar', ax=axes[0], color='black', title='Top 10 Dates With Most Terrorist Attacks')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Count')


# First, we'll extract year and month from date
month = df['date'].dt.to_period('M')

# Extract month name
month_name = month.dt.strftime('%B')

top_months = month_name.value_counts()
top_months \
    .plot(kind='bar', ax=axes[1], color='blue', title='Terrorist Attacks by Month')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Count')

# Show the plots
plt.tight_layout()
plt.show()
```

```python
    # Bar chart for 'Top 5 Countries/ProvinceOrState/Regions With Most Attacks', 'Top 5 Attack/Target Types'

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))

# Country

top_countries = df['country_txt'].value_counts().head(5)
top_countries.plot(kind='bar', ax=axes[0, 0], title='Top 5 Countries With Most Terror Attacks', color='black')
axes[0, 0].set_xlabel('Country')
axes[0, 0].set_ylabel('Count')

# Province

top_provinces = df['provstate'].value_counts().head(5)
top_provinces.plot(kind='bar', ax=axes[0, 1], title='Top 5 Province/States With Most Terror Attacks', color='blue')
axes[0, 1].set_xlabel('Province/State')
axes[0, 1].set_ylabel('Count')

# Attack Type
top_attacks = df['attacktype1_txt'].value_counts().head(5)
top_attacks.plot(kind='bar', ax=axes[1, 1], title='Top 5 Attack Types', color='blue')
axes[1, 1].set_xlabel('Attack Type')
axes[1, 1].set_ylabel('Count')

# Target Type

top_targets = df['targtype1_txt'].value_counts().head(5)
top_targets.plot(kind='bar', ax=axes[1, 0], title='Top 5 Target Types', color='black')
axes[1, 0].set_xlabel('Target Type')
axes[1, 0].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
    # Bar chart for Top 5 Weapon Types/Subtypes Used For Terror Attacks' and 'Top 5 Countries with Most Deaths/Wounded'

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 20))

# Weapon Type

top_weaptypes = df['weaptype1_txt'].value_counts().head(5)
top_weaptypes.plot(kind='bar', ax=axes[1, 1], title='Top 5 Weapon Types Used For Terror Attacks', color='black')
axes[1, 1].set_xlabel('Weapon Type')
axes[1, 1].set_ylabel('Count')

# Weapon Subtype
top_weapsubtypes = df['weapsubtype1_txt'].value_counts().head(5)
top_weapsubtypes.plot(kind='bar', ax=axes[1, 0], title='Top 5 Weapon Subtypes Used For Terror Attacks', color='blue')
axes[1, 0].set_xlabel('Weapon Subtype')
axes[1, 0].set_ylabel('Count')

# Deaths

most_deaths = df.groupby('country_txt')['nkill'].count().sort_values(ascending = False).head(5)

most_deaths.plot(kind='bar', ax=axes[0, 0], title='Top 5 Countries With Most Deaths From Terrorism', color='black')
axes[0, 0].set_xlabel('Country')
axes[0, 0].set_ylabel('Count')

# Wounded

most_wounded = df.groupby('country_txt')['nwound'].count().sort_values(ascending = False).head(5)

most_wounded.plot(kind='bar', ax=axes[0, 1], title='Top 5 Countries With Most Wounded From Terrorism', color='blue')
axes[0, 1].set_xlabel('Country')
axes[0, 1].set_ylabel('Count')
```
